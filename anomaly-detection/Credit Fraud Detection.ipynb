```python
def find_outlier_iqr(df, col):
    q25, q75 = np.percentile(col, 25), np.percentile(col, 75)
    logger.info('Quartile 25: {} | Quartile 75: {}'.format(q25, q75))
    iqr = q75 - q25
    logger.info('IQR: {}'.format(iqr))

    cut_off = iqr * 1.5
    lower, upper = q25 - cut_off, q75 + cut_off
    
    logger.info(f'Cut Off: {cut_off}')
    logger.info(f'Lower: {lower}')
    logger.info(f'Upper: {upper}')

    outliers = [x for x in col if x < lower or x > upper]
    logger.info('Feature V14 Outliers for Fraud Cases: {}'.format(len(outliers)))
    #logger.info('V10 outliers:{}'.format(outliers))
    
    return outliers
```
